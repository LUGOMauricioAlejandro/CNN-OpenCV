```python
#Aclaraciones en Spanish, sin acentos, para esta prueba simple se utiliza una placa de video NVIDIA RTX 3050 ti de 8GB y de 4GB
#importando la libreria para opencv
from __future__ import print_function #Declara la funcion de impresion
import imutils  
import cv2   #Paquete opencv 
```

```python
print("OpenCV Version> {}".format(cv2.__version__) ) #Imprime version de opencv
image = cv2.imread("/home/mauriciolugo/Documentos/tetris_blocks.png") #Carga la imagen local a ser analizada por opencv
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) #Escala de grices 
thresh = cv2.threshold(gray, 255, 255, cv2.THRESH_BINARY_INV)[1] #Valores que toma bit a bit
```

```python
#Chequea el funcionamiento en opencv2 y opencv 4 ----> NVIDIA RTX 3050 ti
if imutils.is_cv2() or imutils.is_cv4(): #Carga el modulo opencv2 o opencv4
    (cnts, _) = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE) #realiza una copia en la memoria RTX
elif imutils.is_cv3(): #Chequea la version de opencv 3 ----> NVIDIA RTX 3050 ti
    (_,cnts, _) = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
cv2.drawContours(image, cnts, -2, (500, 0, 159), 3) 
cv2.imshow("Image", image)
cv2.waitKey(0)
```
